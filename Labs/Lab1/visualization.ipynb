```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("resVis.csv")
```

```python
matplotlib.rcParams['figure.figsize'] = (15,10)
ax = sns.barplot(x='numpublications', y='numauthors', data=df)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of publications',labelpad=15, fontsize=15)
plt.ylabel('Number of authors', labelpad=15, fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax.yaxis.grid(True)
plt.title('Log-normal distribution', fontsize=20)
plt.tight_layout()
plt.savefig("vis.pdf")
plt.show()
```

# We can conclude that the distribution is log-normal
